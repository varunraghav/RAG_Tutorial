```python
from llama_index.core import Document

# Define the persistence directory and PDF file name
PERSIST_DIR = './new_storage_polity'
PDF_FILE = 'Indian_Polity_Laxmi-Kant-6th-Edition-.pdf'

    
# Function to extract text from PDF
def extract_text_from_pdf(pdf_file):
    reader = PdfReader(pdf_file)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

if not os.path.exists(PERSIST_DIR):
    # Extract text from the PDF
    pdf_text = extract_text_from_pdf(PDF_FILE)
    
    # Create a Document object
    document_polity = Document(pdf_text)
    
    # Build the index
    index = VectorStoreIndex.from_documents([document_polity])
    index.storage_context.persist(persist_dir=PERSIST_DIR)
else:
    # Load the index from the existing persistence directory
    storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_context)
```

## Experimenting with Retrievers and Query Engine


##### Default Approach:

index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()
response = query_engine.query("Do veterans have special rules?")
print(response)

##### New Approach

1. Retriever Configuration: Uses VectorIndexRetriever to define how the retrieval should work (e.g., similarity_top_k=10).
2. Response Synthesizer: Configures a response synthesizer (custom logic for processing retrieved data into a response).
3. Query Engine Assembly: Combines components, including optional postprocessors like SimilarityPostprocessor to filter results based on similarity scores.


We support advanced Node filtering and augmentation that can further improve the relevancy of the retrieved Node objects. This can help reduce the time/number of LLM calls/cost or improve response quality.

For example:

1. KeywordNodePostprocessor: filters nodes by required_keywords and exclude_keywords.
2. SimilarityPostprocessor: filters nodes by setting a threshold on the similarity score (thus only supported by embedding-based retrievers)
3. PrevNextNodePostprocessor: augments retrieved Node objects with additional relevant context based on Node relationships.


```python
from llama_index.core import VectorStoreIndex, get_response_synthesizer
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor
```

```python
# build index
index = VectorStoreIndex.from_documents(documents_delta)

# configure retriever
retriever = VectorIndexRetriever(
    index=index,
    similarity_top_k=10,
)

# configure response synthesizer
response_synthesizer = get_response_synthesizer()

# assemble query engine
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    response_synthesizer=response_synthesizer,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.7)],
)

# query
response = query_engine.query("What personal stuff is allowed?")
print(response)
```

```python
# configure retriever
retriever_polity = VectorIndexRetriever(
    index=index_polity,
    similarity_top_k=15,
)

# configure response synthesizer
response_synthesizer = get_response_synthesizer()

# assemble query engine
query_engine = RetrieverQueryEngine(
    retriever=retriever_polity,
    response_synthesizer=response_synthesizer,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.5)],
)

# query
response_polity = query_engine.query("What are the powers of Supreme Court?")
print(response_polity)
```

```python
from llama_index.core.postprocessor import KeywordNodePostprocessor

node_postprocessors = [
    KeywordNodePostprocessor(
        required_keywords=["Ambedkar"], exclude_keywords=["Jennings"]
    )
]
query_engine = RetrieverQueryEngine.from_args(
    retriever_polity, node_postprocessors=node_postprocessors
)
response = query_engine.query("Summarize the criticism of Indian Constitution?")
print(response)
```

#### Query Engine Response Mode

Right now, we support the following options:

1. default: "create and refine" an answer by sequentially going through each retrieved Node; This makes a separate LLM call per Node. Good for more detailed answers.
2. compact: "compact" the prompt during each LLM call by stuffing as many Node text chunks that can fit within the maximum prompt size. If there are too many chunks to stuff in one prompt, "create and refine" an answer by going through multiple prompts.
3. tree_summarize: Given a set of Node objects and the query, recursively construct a tree and return the root node as the response. Good for summarization purposes.
4. no_text: Only runs the retriever to fetch the nodes that would have been sent to the LLM, without actually sending them. Then can be inspected by checking response.source_nodes. The response object is covered in more detail in Section 5.
5. accumulate: Given a set of Node objects and the query, apply the query to each Node text chunk while accumulating the responses into an array. Returns a concatenated string of all responses. Good for when you need to run the same query separately against each text chunk.


```python
query_engine = RetrieverQueryEngine.from_args(
    retriever_full, response_mode="accumulate"
)
response_3 = query_engine.query("Can I bring pets?")
```

# Demonstrating Tools and Memory

```python
from llama_index.core.tools import FunctionTool


# function tools
def multiply(a: float, b: float) -> float:
    """Multiply two numbers and returns the product"""
    return a * b

multiply_tool = FunctionTool.from_defaults(fn=multiply)

def add(a: float, b: float) -> float:
    """Add two numbers and returns the sum"""
    return a + b

add_tool = FunctionTool.from_defaults(fn=add)
```

```python
rag_tool_pol = QueryEngineTool.from_defaults(
    query_engine2_polity,
    name="textbook_rag_tool",
    description="You are a helpful analyst. Reply with helpful and comprehensive answers incorporating data",
)

agent_pol = ReActAgent.from_tools(
    [multiply_tool, add_tool, rag_tool_pol], verbose=True
)

```

```python
response = agent_pol.chat(
    "How many seats were allocated to Cochin in Constituent Assembly ?"
)

print(response)

response = agent_pol.chat(
    "How many seats were allocated to Travancore in Constituent Assembly?"
)

print(response)

response = agent_pol.chat(
    "How much was the total of those two allocations added together? Use a tool to answer any questions."
)

print(response)
```

##### The error occurs because the agent incorrectly skips using the retrieval tool (rag_tool) for a factual query about Travancore, leading to an invalid assumption based on Cochin's result. 

This is technically termed as a tool invocation error or retrieval failure in AI systems using tools, where the agent fails to call the appropriate tool for context-specific tasks.

```python
rag_tool_pol = QueryEngineTool.from_defaults(
    query_engine=query_engine2_polity,
    name="textbook_rag_tool",
    description=(
        "You are a helpful analyst with access to a retrieval tool that fetches specific data from the 'Indian Polity' textbook. "
        "Always use this tool for any queries about historical or specific factual information."
    ),
)


agent_pol = ReActAgent.from_tools(
    tools=[multiply_tool, add_tool, rag_tool_pol],
    verbose=True,
    system_prompt=(
        "You are an intelligent assistant. When answering questions involving factual or historical information, "
        "always use the 'textbook_rag_tool' to fetch specific data from the textbook."
    ),
)
```

```python
response = agent_pol.chat("How many seats were allocated to Travancore in the Constituent Assembly as on December 31, 1947?")
print(response)

response = agent_pol.chat("How many seats were allocated to Cochin in the Constituent Assembly as on December 31, 1947?")
print(response)

response = agent_pol.chat("How much was the total of these two allocations added together? Use a tool to answer any questions.")
print(response)
```
